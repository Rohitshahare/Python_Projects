```python
# importing python libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
# Importing CSV file 

df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# dropping unrelated or empty columns

df.drop(['Status', 'unnamed1'] , axis= 1, inplace= True)
```

```python
df.info()
```

```python
# Checking for any null values in the dataset

pd.isnull(df).sum()
```

### we can see in amount columns there are 12 null values so we have to get rid of this

```python
df.dropna(inplace= True)
```

```python
df.shape
```

 ### changing data type of amount column

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
df.columns
```

```python
# remaning of marital status column
df= df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
# checking min and max values
df[['Age', 'Orders', 'Amount']].describe()
```

# Performing EDA (Exploratory Data Analysis)

## Gender

```python
# plotting a bar chart graph for Gender and it's counthttps://photos.google.com/?pli=1

ag = sns.countplot(x = 'Gender', data = df )

for bars in ag.containers:
    ag.bar_label(bars)
```

```python
gender_amount = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.barplot( x= 'Gender', y= 'Amount', data = gender_amount)
```

```python
gender_amount.head()
```

### From above graph we can see that most of the buyers are Female and even Purchasing power of female is more than male

## Age

```python
tr = sns.countplot(data = df, x= 'Age Group', hue= 'Gender')
for bars in tr.containers: 
    tr.bar_label(bars)
```

```python
# Total amount vs Age Group

age_amount = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)

sns.barplot(x= 'Age Group', y = 'Amount', data= age_amount)
```

### From above graph we can see that most of the buyers are from age group of 26-35

## State

```python
# Total number of Orders form top 10 states

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data= sales_state, x = 'State' , y = 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

### From above graphs we can see that most of the orders and total sales/amount is form the Uttar pradesh, maharashtra and karnataka 

## Marital Status/ Shaadi

```python
ax = sns.countplot(data = df, x = 'Shaadi')

sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state1 = df.groupby(['Shaadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state1, x = 'Shaadi',y= 'Amount', hue='Gender')
```

```python
sales_state1.head()
```

### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ad = sns.countplot(data = df, x = 'Occupation')

for bars in ad.containers:
    ax.bar_label(bars)
```

```python
sales_state2 = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state2, x = 'Occupation',y= 'Amount')
```

### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
sns.set(rc={'figure.figsize':(23,5)})
ab = sns.countplot(data = df, x = 'Product_Category')

for bars in ab.containers:
    ab.bar_label(bars)
```

```python
sales_state3 = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state3, x = 'Product_Category',y= 'Amount')
```

### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
# Top 10 most sold porducts

sales_state4 = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state4, x = 'Product_ID',y= 'Orders')
```

## Conclusion:

### Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
